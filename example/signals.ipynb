```python
from tensorairspace.utils import generate_time_period, convert_tp_to_sec_tp
from tensorairspace.signals.standart import unit_step, sinusoid
from tensorairspace.signals.random import full_random_signal

import matplotlib.pyplot as plt
import numpy as np
```

```python
dt = 0.01
tp = generate_time_period(tn=20)
```

```python
tps = convert_tp_to_sec_tp(tp, dt)
```

```python
tp_unit = unit_step(degree=5, tp=tp, time_step=10, output_rad=False)
```

```python
fig = plt.figure(figsize=(15,4))
plt.clf()
plt.plot(tps,tp_unit, color='green')
plt.legend()
plt.xlabel("t, sec.")
plt.ylabel("Degree")
plt.grid(True)
```

```python
tp_sinusoid = sinusoid(tp=tp, amplitude=0.01, frequency=10)
```

```python
fig = plt.figure(figsize=(15,4))
plt.clf()
plt.plot(tps,tp_sinusoid, color='green')
plt.legend()
plt.xlabel("t, sec.")
plt.ylabel("Degree")
plt.grid(True)
```

```python
signal =  full_random_signal(0,0.01,20, (-0.5, 0.5), (-0.5, 0.5))
```

```python
fig = plt.figure(figsize=(15,4))
plt.clf()
plt.plot(tps[:2001],signal, color='green')
plt.legend()
plt.xlabel("t, sec.")
plt.ylabel("Degree")
plt.grid(True)
```
