```python

env = env
num_actions = env.action_space.n
model = Model(num_actions)
target_model = Model(num_actions)
```

```python
agent = PERAgent(model, target_model, env, train_nums=1000000)
```

```python
env.action_space.sample()
```

```python
agent.train()
# test after
```
